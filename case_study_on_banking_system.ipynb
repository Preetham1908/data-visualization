```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age Group': ['<21', '21-24', '24-27', '27-30', '30-33','33-36','36-39','39-42','42-45','45-48','48-51','51-54','54-57','57-60','>60'],
    'Total Number of Loans': [9, 310, 511, 4000, 4568,5698,8209,8117,9000,7600,6000,4000,2000,788,6],
    'Number of Bad Loans': [2,200,20,172,169,1888,197,211,216,152,84,640,26,9,0],
    'Number of Good Loans':[7,296,491,3828,4399,5510,802,7906,8784,7448,5916,3936,1974,779,6],
    '% of Bad Loans':['22.2%','4.5%','3.9%','4.3%','3.7%','3.3%','2.4%','2.0%','1.4%','1.6%','1.3%','1.1%','0.0%','5.8%','7.3%'],
    '% of Good Loans':['77.8%','95.5%','96.1%','95.7%','96.3%','95.8%','93.7%','92.9%','96.9%','99.5%','89.7%','45.9%','56.7%','98.4%','43.4%'],
    'Age': [20, 22.5, 25.5, 28.5, 31.5, 34.5, 37.5, 40.5, 43.5, 46.5, 49.5, 52.5, 55.5, 58.5, 65]
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)

combined_df = pd.DataFrame({
    'Age Group': df['Age Group'],
    '% of Good Loans': df['% of Good Loans'].str.rstrip('%').astype(float),
    '% of Bad Loans': df['% of Bad Loans'].str.rstrip('%').astype(float)
})

plt.figure(figsize=(10, 6))

plt.bar(combined_df['Age Group'], combined_df['% of Good Loans'], color='b', )
plt.bar(combined_df['Age Group'], combined_df['% of Bad Loans'], color='y', bottom=combined_df['% of Good Loans'])
plt.ylabel('% of Loans')
plt.title('Number of Loans by Age Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```
